```python
import pymongo
import random
```

```python
client = pymongo.MongoClient("mongodb://10.1.1.9:27017,10.1.1.10:27018,10.1.1.11:27019")
db = client['Neo4Food']
coll = db['Restaurants']
```

```python
result = coll.find({})

zips = {}

for x in result:
    zip = x['zip_code']

    if zip not in zips:
        zips[zip] = 1
    else:
        zips[zip] += 1
```

```python
total = 0
for k in zips:
    total += zips[k]
total
```

```python
enough = []
notEnough = []
rome = []

for k in zips:
    # Se e' di Roma lo lascio dove sta
    if k.startswith("00",0,3):
        rome.append(k)
        continue
    # Se e' meno di 10 lo metto in notEnough
    if zips[k] < 10:
        notEnough.append(k)
    else:
        enough.append(k)
        
```

```python
print(f"Zips in enough: {len(enough)}\nZips in notEnough: {len(notEnough)}")
```

```python
enoughCount = 0
notEnoughCount = 0
romeCount = 0

for k in zips:
    if k in enough:
        enoughCount += zips[k]
    else:
        if k in notEnough:
            notEnoughCount += zips[k]
        else:
            romeCount += zips[k]

print(f"enough count: {enoughCount}\nNot enough count: {notEnoughCount}\nRome count: {romeCount}\n\nSum compared to old total:(S: {enoughCount+notEnoughCount+romeCount} - T: {total})")
```

```python
cursor = coll.find()

for x in cursor:
    id = x['_id']
    if x['zip_code'] in notEnough:
        pickOne = random.choice(enough)
        coll.update_one({"_id": id},{"$set": {"zip_code": pickOne}})
```
